```python
import cv2
import imutils
from imutils.object_detection import non_max_suppression
import numpy as np

hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
image = cv2.imread("ped3.jpg")
image = imutils.resize(image, width=min(1000, image.shape[1]))
orig = image.copy()

# detect people in the image
(rects, weights) = hog.detectMultiScale(image, winStride=(5, 5),padding=(8, 8), scale=1.035)


rects = np.array([[x, y, x + w, y + h] for (x, y, w, h) in rects])
pick = non_max_suppression(rects, probs=None, overlapThresh=0.65)



centers=[]
r = 0
for (xA, yA, xB, yB) in pick:
    r += 1
    cv2.rectangle(image, (xA, yA), (xB, yB), (0, 255, 0), 2)
    cv2.putText(image, f'P{r}', (xA, yA), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
    xC=xA+int((xB-xA)/2)
    yC=yA+int((yB-yA)/2)
    centers.append((xC,yC))
    
cv2.putText(image, f'Total No. of Person:{r}', (0, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 0), 2)    

nears=[]
k=min(1000, image.shape[1])
print(k)
print(image.shape[0])
minDistance=image.shape[1]//6
if( image.shape[0]>600):
    minDistance=image.shape[1]//4

arr=[ ] 

for k in range(0, r):
    arr.append(0)
m=0
for c,p1 in enumerate(centers):
    for b,p2 in enumerate(centers[c+1:]):
        
        print(minDistance)
        print(np.array(p2))
        print(np.array(p1))
        v=np.linalg.norm(np.array(p2)-np.array(p1))
        if(image.shape[0]>600):
            v=(np.linalg.norm(np.array(p2)-np.array(p1)))/2
        print(v)
        if (v)<minDistance:
            if abs(p1[1] - p2[1])< image.shape[0]*0.1:
                nears.append((c,b+c+1))
                m+=1
                #print(arr[i]," ",arr[j])
                #break;

                
print(nears)

c=0
for (i,j) in nears:
    (xA, yA, xB, yB) = pick[i]
    cv2.rectangle(image, (xA, yA), (xB, yB), (0, 0, 255), 2)
    c=c+1
    if(arr[i]==0):
        arr[i]+=1
                    #print('cc')
              

for (i,j) in nears:
    (xA, yA, xB, yB) = pick[j]
    cv2.rectangle(image, (xA, yA), (xB, yB), (0, 0, 255), 2)
    c+=1
    if(arr[j]==0):
        arr[j]+=1
    

print(arr)

d=0
for k in arr:
    if k==1:
        d+=1

cv2.putText(image, f'No. of Person violating rule:{d}', (0, 20), cv2.FONT_HERSHEY_SIMPLEX,0.6, (0, 255, 255), 2)   
   


cv2.imshow("Alerts", image)
if cv2.waitKey(0):
    cv2.destroyAllWindows()
```
