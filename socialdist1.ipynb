```python
import cv2
import imutils
from imutils.object_detection import non_max_suppression
import numpy as np

hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())


cap = cv2.VideoCapture('video1.mp4') 
while cap.isOpened(): 
   
    ret, image = cap.read() 
    if ret: 
        image = imutils.resize(image, width=min(900, image.shape[1]))
        orig = image.copy()

        (rects, weights) = hog.detectMultiScale(image, winStride=(4,4), padding=(8, 8), scale=1.03)


        rects = np.array([[x, y, x + w, y + h] for (x, y, w, h) in rects])
        picture = non_max_suppression(rects, probs=None, overlapThresh=0.7)
        

        centers=[]
        r = 0
        for (xA, yA, xB, yB) in picture:
            #if (xB-xA)*(yB-yA)>5000:
             #   continue
            #yB=yB-20
            r += 1
            cv2.rectangle(image, (xA, yA), (xB, yB), (0, 255, 0), 2)
            cv2.putText(image, f'P{r}', (xA, yA), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
            xC=xA+int((xB-xA)/2)
            yC=yA+int((yB-yA)/2)
            yD=yA+int((yB-yA-2)/2)
            centers.append((xC,yC))
        
        arr=[ ] 
        for k in range(0, r):
            arr.append(0)    
        cv2.putText(image, f'Total No. of Person:{r}', (0, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 0), 2)    

        nears=[]
        minDistance=image.shape[1]//9

        for i,p1 in enumerate(centers):
            for j,p2 in enumerate(centers[i+1:]):
                if (np.linalg.norm(np.array(p2)-np.array(p1)))<minDistance:
                    if abs(p1[1] - p2[1])< image.shape[0]*0.1:
                        nears.append((i,j+i+1))


           
        c=0;
        for (i,j) in nears:
            #yB=yB-20
            c+=1
            (xA, yA, xB, yB) = picture[i]
            cv2.rectangle(image, (xA, yA), (xB, yB), (0, 0, 255), 2)
            (xA, yA, xB, yB) = picture[j]
            cv2.rectangle(image, (xA, yA), (xB, yB), (0, 0, 255), 2)
            if(arr[i]==0):
                arr[i]+=1
            if(arr[j]==0):
                arr[j]+=1
        k=-1;
        for (xA, yA, xB, yB) in picture:
            k+=1;
            print((yB-yA)*(xB-xA))
            if ((yB-yA)*(xB-xA))>20000:
                cv2.rectangle(image, (xA, yA), (xB, yB), (0, 0, 255), 2)
                if(arr[k]==0):
                    arr[k]+=1
        d=0
        for k in arr:
            if k==1:
                d+=1
            
        cv2.putText(image, f'No. of Person violating rule:{d}', (0, 20), cv2.FONT_HERSHEY_SIMPLEX,0.6, (0, 255, 255), 2)   
        cv2.imshow("Image", image) 
        if cv2.waitKey(25) & 0xFF == ord('q'): 
            break
    else: 
        break

cap.release() 
cv2.destroyAllWindows()

    
    
```
